### TASK 1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


<p>You will require:-

```python
# your code here
average_sales_per_year = df.groupby('Year')['Automobile_Sales'].sum()
plt.figure(figsize=(8, 6))
average_sales_per_year.plot(kind='line', marker='')
plt.title('Automobile Sales per Year')
plt.xlabel('Year')
plt.ylabel('Total yearly sale')
plt.xticks([1980,1981,1982,1983,1990,1991,1992,2000,2001,2002,2008,2009,2020])#顯示年分
plt.tick_params(axis='x', rotation=90)

y_min, y_max = plt.ylim()
plt.annotate(
    'Recession 1980-1982',
    xy=(1981.5, y_max * 0.15),
    xytext=(1983, y_max * 0.9),
    arrowprops=dict(facecolor='blue', shrink=0.08),
    fontsize=10,
    color='black'
)
plt.annotate(
    'Recession 2000-2001',
    xy=(2001, y_max * 0.15),
    xytext=(2002, y_max * 0.9),
    arrowprops=dict(facecolor='blue', shrink=0.08),
    fontsize=10,
    color='black'
)
plt.show()
```

### TASK 4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


```python
# your code here

average_gdp = df.groupby(['Year', 'Recession'], as_index=False)['GDP'].mean()

fig = plt.figure(figsize=(10, 6))

ax1 = fig.add_subplot(1, 2, 1)
recession_data = average_gdp[average_gdp['Recession'] == 1]
ax1.plot(recession_data['Year'], recession_data['GDP'], marker='', color='darkblue')
ax1.set_title('GDP Variation During Recession')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_ylim(20, 60)
ax1.set_xticks(range(1980, 2024, 10))
ax1.grid()

ax2 = fig.add_subplot(1, 2, 2)
non_recession_data = average_gdp[average_gdp['Recession'] == 0]
ax2.plot(non_recession_data['Year'], non_recession_data['GDP'], marker='', color='darkblue')
ax2.set_title('GDP Variation During Non-Recession')
ax2.set_xlabel('Year')
ax2.set_ylabel('GDP')
ax2.set_ylim(20, 65)
ax2.set_xticks(range(1980, 2024, 10))
ax2.grid()

plt.tight_layout()
plt.suptitle('Comparing GDP in Recession and Non-recession period')
plt.subplots_adjust(top=0.9)
plt.show()
```

### TASK 5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<p>You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>


```python
non_recession_data = df[df['Recession'] == 0]#沒有Recession的年

monthly_avg_sales = non_recession_data.groupby('Month').agg({
    'Automobile_Sales': 'mean',
    'Seasonality_Weight': 'mean' #平均權重
}).reset_index()

#月份字典
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

monthly_avg_sales['Month'] = monthly_avg_sales['Month'].map(month_mapping)#月份轉數字

#氣泡圖
plt.figure(figsize=(8, 6))
plt.scatter(monthly_avg_sales['Month'], 
            monthly_avg_sales['Automobile_Sales'], 
            s=monthly_avg_sales['Seasonality_Weight'] * 100,
            alpha=0.6, 
            c='blue', 
            edgecolors='w',
            linewidth=0.5)
for weight in [0.25,0.50,0.75,1.00,1.25,1.50]:
    plt.scatter([], [], s=weight * 100, color='blue', alpha=0.6, edgecolors='w', label=f'{weight}')
plt.legend(title="Seasonality_Weight", fontsize=10, title_fontsize=12, loc='lower right')

plt.title('Seasonality Impact on Automobile Sales', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Automobile Sales', fontsize=14)
plt.grid()
plt.xticks(range(1, 13), 
           ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
            'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
