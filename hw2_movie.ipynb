```python
import pandas as pd

#建立關係
def relation(df):
    graph = {}#存每個演員合作關係的字典
    for _, row in df.iterrows():
        actors = [actor.strip() for actor in row['Actors'].split('|')]#當前行的演員
        for actor in actors:
            if actor not in graph:#若actor沒在關係圖中，則創建新集合
                graph[actor] = set()
            for co_actor in actors:
                if co_actor != actor:#防止自己跟自己合作
                    graph[actor].add(co_actor)
    return graph

def collaborator(graph, start_actor):
    visited = set()#防止重複紀錄
    queue = [start_actor]#對列起始是要搜尋的演員
    colla = set()

    while queue:
        actor = queue.pop(0)#從最前面取一個
        if actor not in visited:
            visited.add(actor)
            colla.add(actor)
            queue.extend(graph[actor] - visited)#沒訪問過的加到visited

    colla.remove(start_actor)#扣掉自己
    return colla

relation = relation(df)
collaborators = collaborator(relation,"Johnny Depp")
collaborators
```
